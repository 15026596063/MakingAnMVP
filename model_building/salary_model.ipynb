```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# Read the csv file into a pandas DataFrame
salary_csv = os.path.join('..', 'data', 'salary_model_data.csv')

salary_stat_df = pd.read_csv(salary_csv)
salary_stat_df = salary_stat_df.rename(columns={'\n\t\t\t\t\t\t\t2017/18\t\t\t\t\t\t': '2018_Salary'})
salary_stat_df.head(10)
```

```python
# drop data not used in model
model_data = salary_stat_df.iloc[:, 3:]
model_data.head(10)
```

```python
# prepare stats data
X = model_data.iloc[:, :-1]

# transform MVP column to binary
X = pd.get_dummies(X)
```

```python
# prepare salary data
y = model_data[['2018_Salary']]

# convert salaries to floats
y = y.replace('[\$,]', '', regex=True).astype(float)
```

```python
# confirm stat and salary data shape
print(X.shape, y.shape)
```

```python
# Use train_test_split to create training and testing data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# # create standard scaler
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
# Create the model using LinearRegression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data
model.fit(X_train_scaled, y_train_scaled)
training_score = model.score(X_train_scaled, y_train_scaled)
testing_score = model.score(X_test_scaled, y_test_scaled)

# model.fit(X_train, y_train)
# training_score = model.score(X_train, y_train)
# testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
coefficients = model.coef_[0]
coefficients
```

```python
from sklearn.feature_selection import f_regression

f_scores, p_values = f_regression(X_train_scaled, y_train_scaled, center=True)
```

```python
stats_list = X_train.columns
stats_list
```

```python
stat_dict = {
    'statistic': [],
    'f_score': [],
    'p_value': [],
    'coefficient': []
}

for stat, f_score, p_value, coef in zip(stats_list, f_scores, p_values, model.coef_[0]):
    
    stat_dict['statistic'].append(stat)
    stat_dict['coefficient'].append(coef)
    stat_dict['f_score'].append(f_score)
    stat_dict['p_value'].append(p_value)
```

```python
stat_df = pd.DataFrame(stat_dict)
stat_df
```

```python
stat_df.to_csv('model_stats.csv')
```
