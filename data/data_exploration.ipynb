```python
import pandas as pd
import os
import numpy as np
```

```python
# url for player salaries
url = 'http://hoopshype.com/salaries/players/'
```

```python
# dataframe of salaries
salaries_df = pd.read_html(url)
salaries_df = salaries_df[0]
salaries_df.head(10)
```

```python
# trimmed down to 2018 season salaries
salaries_df = salaries_df.iloc[:,1:3]
print(len(salaries_df))
salaries_df.head(10)
```

```python
# player stats data
players_csv = os.path.join('nba-players-stats-since-1950', 'Players.csv')
stats_csv = os.path.join('final_seasonstats_.csv')
raw_stats_csv = os.path.join('nba-players-stats-since-1950', 'Seasons_Stats.csv')
```

```python
years_played_df = pd.read_csv(raw_stats_csv)
years_played_df.head(10)
```

```python
years_played = years_played_df.iloc[:, 1:4]
years_played.head(10)
```

```python
years_played_agg = years_played.groupby( [ "Year", "Player"] ).size().to_frame(name = 'count').reset_index()
years_played_agg.head(10)
```

```python
years_played_agg = years_played_agg.groupby('Player').size().to_frame(name = 'years_played').reset_index()
years_played_agg.head(10)
```

```python
# dataframe with player data
players_df = pd.read_csv(players_csv)
players_df.head()
```

```python
edited_players = players_df.iloc[:, 1:4]
edited_players.head(10)
```

```python
# dataframe with player stats data
stats_df = pd.read_csv(stats_csv)
stats_df.head()
```

```python
edited_stats = stats_df.loc[stats_df['Year'] == 2017]
print(len(edited_stats))
edited_stats.head()
```

```python
complete_stats = pd.merge(edited_stats, edited_players, on='Player', how='left')
complete_stats.head(10)
```

```python
complete_stats = pd.merge(complete_stats, years_played_agg, on='Player', how='left')
complete_stats.head(10)
```

```python
final_dataset = pd.merge(complete_stats, salaries_df, on='Player', how='left')
print(len(final_dataset))
final_dataset.head(10)
```

```python
null_cols = final_dataset[final_dataset.isnull().any(axis=1)]
print(len(null_cols))
null_cols.head()
```

```python
clean_data = final_dataset.dropna()
print(len(clean_data))
clean_data.head(10)
```

```python
# save data
clean_data.to_csv('salary_model_data.csv')
```
